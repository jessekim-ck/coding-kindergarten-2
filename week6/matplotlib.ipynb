```python
# 이미지 불러오기
img = plt.imread("data/cbc.jpeg")
```

```python
# 이미지 보여주기
plt.figure(figsize=(30, 10))  # 필요 없음 - 이미지 보여주는 사이즈 조정하는 코드
plt.imshow(img)
```

```python
# 이미지는 사실 numpy array 였다!
img
```

```python
img.shape  # 1334 x 750 크기의 이미지, RGB 채널
```

```python
plt.figure(figsize=(30, 10))

# 이미지 반전시키기
new_img = np.zeros_like(img)  # 같은 크기의 빈 array(까만색 이미지)를 하나 만들고,
new_img[:, :, 1] = img[:, :, 0]  # R 채널을 G 채널에,
new_img[:, :, 2] = img[:, :, 1]  # G 채널을 B 채널에,
new_img[:, :, 0] = img[:, :, 2]  # B 채널을 R 채널에 넣으면
plt.imshow(new_img)  # 식욕 억제 이미지 탄생!
```
